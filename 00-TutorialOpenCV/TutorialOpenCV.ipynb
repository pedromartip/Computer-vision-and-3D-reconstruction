### Slicing

Slicing operator is a very useful operation for working with matrices and vectors. This is a python native operation when it is applied to lists, but it is extended to numpy matrices and vectors, where they are very efficient.

This operation is applied to an existing vector/matrix to create a new instance from the original (https://www.tutorialsteacher.com/python/slice-method). The syntax of the operation is as follows:
```python
arr[start:stop:step]
```
Where:
* arr: Vector where we are applying the operation.
* start: Index of the arr element which we want to be the first element of the new vector. By default is 0.
* stop: Index of the arr element which we want to be the last element of the new vector.
* step: How we want to iterate arr to create the new vector. By default is 1
